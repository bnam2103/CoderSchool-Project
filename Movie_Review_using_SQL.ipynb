### **Question 4 (SQL - 9 pts)**
- For each movie, return the title and the 'rating spread', that is, the **difference between highest and lowest ratings** given to that movie. 
- Sort by rating spread from **highest to lowest, then by movie title ascendingly**.

```python
#@title **Your result should look like this** 
```

```python
query = ''' SELECT title, (MAX(stars)-MIN(stars)) as Spread
FROM Rating
LEFT JOIN Movie USING(mID)
GROUP BY title
ORDER BY Spread DESC
'''

pd.read_sql_query(query, conn)
```

- Find the titles of all movies **not** reviewed by **Elizabeth Thomas**.

**Your result should look like this** 

|    | title                   |

|  1 | E.T.                    |
|  2 | Raiders of the Lost Ark |
|  3 | Titanic                 |
|  4 | Twilight                |
|  5 | Wanted                  |

```python
query = '''SELECT Title 
FROM Movie
    WHERE mID NOT IN
      (
        SELECT RA.mID
        FROM Rating RA
            JOIN Reviewer R USING (rID)
            WHERE R.name = 'Elizabeth Thomas'
      )
    ORDER BY Title
'''

pd.read_sql_query(query, conn)
```

### **Question 9 (SQL - 9 pts)**
- For each director, return the director's name together with the title(s) of the movie(s) they directed **that received the highest rating among all of their movies**, and the value of that rating. 
- Ignore movies whose director is NULL.

```python
#@title **Your result should look like this** 
```

```python
query = ''' WITH tmp as
(
SELECT MAX(stars) as star_max, mID
FROM Rating
GROUP BY mID
)
SELECT director as name, title, MAX(star_max) as stars 
FROM movie
JOIN tmp USING (mID)
WHERE name IS NOT NULL
GROUP BY name
ORDER BY name ASC

'''

pd.read_sql_query(query, conn)
```

Hint: JOIN can be used without ON or USING () to join using Index

```python
#@title **Your result should look like this** 
```

```python
query = ''' 
  WITH tmp as
(  
  SELECT AVG(stars) as stars, '1980>' as year
   FROM Movie
   LEFT JOIN Rating USING (mID)
   WHERE Movie.year >= 1980
   
   UNION
  SELECT AVG(stars) as stars, '1980<' as year
    FRom movie
    JOIN rating USING (mID)
    WHERE Movie. year <1980
)
  SELECT MAX(stars)- MIN(stars) as diff
    FROM tmp
'''

pd.read_sql_query(query, conn)
```
